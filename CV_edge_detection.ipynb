# Τί είναι η ανίχνευση ακμών;

Η ψηφιακή επεξεργασία εικόνας περιλαμβάνει ένα σύνολο αλγορίθμων για την επεξεργασία, ανάλυση και χειρισμό των ψηφιακών δεδομένων εικόνας ή βίντεο. Οι αλγόριθμοι αυτοί δέχονται ως είσοδο εικόνες/βίντεο και παράγουν ως έξοδο εικόνες/βίντεο και μας βοηθάνε να εξάγουμε και να χρησιμοποιήσουμε τις πληροφορίες που προέρχονται από το συγκεκριμένο ψηφιακό περιεχόμενο. <br> Η ανίχνευση ακμών είναι μια τεχνική που χρησιμοποιούμε στην επεξεργασίας εικόνας για να <b>εντοπίσουμε σημεία της εικόνας τα οποία παρουσιάζουν ασυνέχειες</b>, δηλαδή έντονες αλλαγές στη φωτεινότητα της εικόνας. Μία ακμή σε μια εικόνα, θα μπορούσαμε να πούμε ότι χαρακτηρίζεται από μια <b>σημαντική τοπική αλλαγή στην ένταση της εικόνας (από pixel σε pixel)</b>.<br> Παράδειγμα: 

```python
from matplotlib import pyplot as plt
# import the cv2 library
import cv2

# The function cv2.imread() is used to read an image.
img = cv2.imread('imgs/letter_A.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# # The function cv2.imshow() is used to display an image in a window.
# cv2.imshow('graycsale image',img_grayscale)
plt.imshow(gray)
plt.title('image of letter "A"')
plt.show()
```
